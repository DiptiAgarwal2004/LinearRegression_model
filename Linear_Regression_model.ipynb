```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Sample dataset
data = pd.DataFrame({'x1': [1, 2, 3, 4, 5],
                      'x2': [3, 5, 7, 1, 4],
                      'y': [8, 13, 18, 9, 15]})

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(data[['x1', 'x2']], data['y'])

# Access the coefficients and intercept
coefs = model.coef_  # Coefficients for each feature
intercept = model.intercept_  # Intercept term

print("Coefficients:", coefs)
print("Intercept:", intercept)
```

###2. Make predictions using the data on unknown data points

```python
# Split the data into training and testing sets(80% training, 20% testing)

from sklearn.model_selection import train_test_split #train and test

X = ds[['temp', 'humidity']]
y = ds['visibility']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression   #build ML model

'''Training'''
# Fit the linear regression model on the training data
model = LinearRegression()
model.fit(X_train, y_train)


'''Testing'''
# Make predictions on the testing data
y_pred = model.predict(X_test)
print("Prediction value: ",y_pred)

# Evaluate the model performance using R-squared
r_squared = model.score(X_test, y_test)
print('R-squared:', r_squared)

# Calculate the mean squared error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("Mean squared error: ", mse)
```
